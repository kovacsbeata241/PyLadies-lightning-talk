```python
#új tábla megnyitása
df = pd.read_csv('Budapesti maximum hőmérsékletek.csv')
df
```

```python
#oszlopnevek 
df.columns
```

```python
#felesleges oszlopok "eldobása"
df.drop(['name', 'tempmin', 'temp', 'feelslikemax', 'feelslikemin', 'feelslike', 'dew', 'humidity', 'precip', 'precipprob', 'precipcover', 'preciptype', 'snow', 'snowdepth', 'windgust', 'windspeed', 'winddir', 'sealevelpressure', 'cloudcover', 'visibility', 'solarradiation', 'solarenergy', 'uvindex', 'severerisk', 'sunrise', 'sunset', 'moonphase', 'conditions', 'description', 'icon', 'stations'], axis = 1, inplace = True)
df
```

```python
#dátum konvertálása
df['datetime'] = pd.to_datetime(df['datetime'])
df
```

```python
#színkód oszlop létrehozása a hőmérsékletek 8 részre osztásával
df['colorcode'] = pd.cut(df['tempmax'], bins=8, labels=range(1, 9))
df
```

```python
#szín és fonal segédtábla létrehozása
yarn_assist = pd.DataFrame(data={'colorcode': [1, 2, 3, 4, 5, 6, 7, 8],
                       'color': ['LightBlue', 'RoyalBlue', 'MidnightBlue', 'MediumSeaGreen', 'DarkSeaGreen', 'OldLace', 'Pink', 'MediumOrchid'],
                       'name': ['Jégkék', 'Denimkék', 'Indigó', 'Ködös erdő', 'Pisztácia', 'Fehér köd', 'Rózsaszín', 'Ametiszt'],
                       'yarn': ['DROPS Flora Uni Colour 14', 'DROPS Flora Uni Colour 13', 'DROPS Flora Uni Colour 10', 'DROPS Flora Uni Colour 23', 
                       'DROPS Flora Uni Colour 16', 'DROPS Flora Uni Colour 22', 'DROPS Flora Uni Colour 21', 'DROPS Flora Uni Colour 09']})
yarn_assist
```

```python
#új oszlop a színekkel (egyesítés a segédtáblával)
df = df.merge(yarn_assist, how = 'left', on = ['colorcode'])
df
```

```python
start_date = df.loc[:,'datetime'].min(axis = 0)
start_date
```

```python
cut_date = start_date + datetime.timedelta(days = 365)
cut_date
```

```python
#év oszlop létrehozása
df['year'] = np.where(df['datetime'] < cut_date, 1, 2)
df
```

```python
#sor oszlop létrehozása, 1 év után újrakezdődik
df['row'] = (df.index % 365) + 1
df
```

```python
#szökőév miatti plusz sor elhagyása
df = df.drop(df.index[730])
df
```

```python
#fonalmennyiség számításhoz külön tábla - egyes színek megszámolása
yarn = df['colorcode'].value_counts()
yarn = yarn.reset_index()
yarn.columns = ['colorcode', 'Count']
yarn = yarn.merge(yarn_assist, how = 'left', on = ['colorcode'])
yarn = yarn.sort_values(by = 'colorcode')
yarn
```

### Mérés: 21 szem 30,5cm fonal, 120 szem van egy sorban, sorok két végén "lelógó fonal" soronként 30 cm 

### 1 gombolyag (50g) 210 méter fonalat tartalmaz

```python
#átváltási ráta kiszámítása - 1 sor hány cm fonal
convertion_rate = 120 / 21 * 30.5 + 30
convertion_rate
```

```python
# fonalszükséglet kiszámítása
yarn['yarn_m'] = yarn['Count'] * convertion_rate / 100 #fonalszükséglet méterben
yarn
```

```python
yarn['ball'] = yarn['yarn_m'] / 210 #gombolyagszám
yarn
```

```python
yarn['ball_whole'] = np.ceil(yarn['yarn_m'] / 210).astype(int) #gombolyagszám felfelé egészre kerekítve
yarn
```

```python
#poncsó súlyának becslése g-ban
weight = yarn.loc[:,['ball']].sum(axis = 0) * 50
weight
```
